# PART 1 DATAFRAME FOR TORONTO

```python
import pandas as pd

url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df=pd.read_html(url, header=0)[0]

df.head()
```

```python
df.shape
```

```python
df = df[df.Borough!='Not assigned']
df.head()
```
